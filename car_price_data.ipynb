# **Bar Plot**

```python
def bar_plot(variable):
    var = df[variable]                  #comes from dataset
    varValue=var.value_counts()
    plt.figure(figsize=(20,3))
    plt.bar(varValue.index,varValue)                       #why varValue because we need numerical value
    plt.xticks(varValue.index,varValue.index.values)
    plt.ylabel("Frequency")
    plt.title(variable)
    plt.show()
    print("{}: \n {}".format(variable,varValue))
```

__Bar Plot of each Columns__

```python
columns_plot=['Brand', 'Model', 'Year', 'Engine_Size', 'Fuel_Type', 'Transmission',
       'Mileage', 'Doors', 'Owner_Count', 'Price']
for i in columns_plot:
    bar_plot(i)
```

__Hist Plot__

```python
columns_plot = ['Brand', 'Model', 'Year', 'Engine_Size', 'Fuel_Type', 'Transmission',
       'Mileage', 'Doors', 'Owner_Count', 'Price']

for i in columns_plot:
    plt.figure(figsize=(9, 3))  # Set the figure size
    sns.histplot(data=df, x=i)  # Create a histogram plot
    plt.title(f'Histogram of {i}')  # Add a title to the plot
    plt.savefig(f'{i}_histplot.png')  # Save the plot as an image file
    plt.close() 
```
